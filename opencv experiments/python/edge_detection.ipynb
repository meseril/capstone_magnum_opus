```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("saved_images/US-Type-Coins-Major-56a1794e5f9b58b7d0bfa597.jpg")
resized_img = cv2.resize(img, (600,400))
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

canny_img = cv2.Canny(resized_img, 100, 200)

plt.subplot(121), plt.imshow(rgb_img), plt.title("Original Image")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(canny_img), plt.title("Detected edges")
plt.xticks([]), plt.yticks([])

plt.show()
```
