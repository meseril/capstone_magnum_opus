```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("./images/flemingo.jpg", 0)
resize_img = cv2.resize(img, (400,400))


histogram = cv2.calcHist([resize_img],[0], None, [255],[0,255])

# cumulative distribution function
cdf = histogram.cumsum()
cvd_normalized = cdf * float(histogram.max()) / cdf.max()

plt.plot(cvd_normalized, color="b")
plt.hist(resize_img.flatten(), 255,[0,255], color="r")
plt.xlim([0,255])
plt.legend(("CDF","Histogram"))

plt.show()
 
hist_eq = cv2.equalizeHist(resize_img)
cv2.imshow("Histogram Equalizition",hist_eq)
cv2.waitKey()
cv2.destroyAllWindows()

```
