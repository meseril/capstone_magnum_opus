```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import imutils
```

```python
img = cv2.imread("./saved_images/coins.jpg")
img = cv2.resize(img, (600,480))

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

thres = cv2.threshold(gray_img, 187, 255, cv2.THRESH_BINARY_INV)[1]
contours = cv2.findContours(thres.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contours = imutils.grab_contours(contours)
output = img.copy()

cv2.drawContours(output, contours, -1, (40,50,200), 3)
text = "{} objects found".format(len(contours))
cv2.putText(output, text, (15,15), cv2.FONT_HERSHEY_SIMPLEX, 0.7,(0,0,0),2)

plt.subplot(221), plt.imshow(rgb_img), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(gray_img, cmap="Greys_r"), plt.title("Gray")
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(thres, cmap="Greys_r"), plt.title("After Thresholding")
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(output), plt.title(text)
plt.xticks([]), plt.yticks([])

plt.show()
```
